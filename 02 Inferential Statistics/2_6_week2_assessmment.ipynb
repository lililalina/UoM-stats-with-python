### Average bedtime confidence interval for napping and non napping toddlers
Create two 95% confidence intervals for the average bedtime, one for toddler who nap and one for toddlers who don't.

```python
# Isolate the column napping and night bedtime
df_bedtime = df[['napping', 'night bedtime']]
df_bedtime = df_bedtime.groupby(['napping']).agg({'night bedtime':[ np.mean, np.std, np.size]})
df_bedtime
```

```python
# Average bedtime for toddlers who nap and its 95% CI
n1 = df_bedtime.iloc[1,2]
nap_mean_bedtime = df_bedtime.iloc[1,0]
nap_s_bedtime = df_bedtime.iloc[1,1] 
nap_se_mean_bedtime = nap_s_bedtime/(np.sqrt(n1))

# t value for n = 15, CI 95%
nap_t_star = t.ppf(1- ((100-95)/2/100), 14)

# Calculate the lcb and ucb
lcb_bedtime_nap = nap_mean_bedtime - nap_se_mean_bedtime*nap_t_star
ucb_bedtime_nap = nap_mean_bedtime + nap_se_mean_bedtime*nap_t_star

print('Average bedtime for toddlers who nap:', nap_mean_bedtime, "Â±", nap_se_mean_bedtime)
print('Lower bound:', lcb_bedtime_nap, "Upper bound:", ucb_bedtime_nap)
```

```python
# Average bedtime for toddlers who do not nap and its 95% CI
n2 = df_bedtime.iloc[0,2]
no_nap_mean_bedtime = df_bedtime.iloc[0,0]
no_nap_s_bedtime = df_bedtime.iloc[0,1] 
no_nap_se_mean_bedtime = no_nap_s_bedtime/(np.sqrt(n2))

# t value for n = 5, CI 95%
no_nap_t_star = t.ppf(1- ((100-95)/2/100), 4)

# Calculate the lcb and ucb
lcb_bedtime_nonap = no_nap_mean_bedtime - no_nap_se_mean_bedtime*no_nap_t_star
ucb_bedtime_nonap = no_nap_mean_bedtime + no_nap_se_mean_bedtime*no_nap_t_star

print('Average bedtime for toddlers who do not nap:', no_nap_mean_bedtime, "Â±",no_nap_se_mean_bedtime)
print('Lower bound:', lcb_bedtime_nonap, "Upper bound:", ucb_bedtime_nonap)
```

#### Notes for t value
Example: to find the $t^*$ for a 90% confidence interval, we want $t^*$ such that 90% of the density of the $t$ distribution lies between $-t^*$ and $t^*$.

Or in other words if $X \sim t(df)$:

P($-t^*$ < X < $t^*$) = .90

Which, because the $t$ distribution is symmetric, is equivalent to finding $t^*$ such that:  

P(X < $t^*$) = .95

(0.95 = 1 - (1 - confidence) / 2 = 1 - 0.1 / 2 = 1 - 0.05)

So the $t^*$ for a 90% confidence interval, and lets say df=10, will be:

t_star = t.ppf(.95, df=10)
